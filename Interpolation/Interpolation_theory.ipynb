# Interpolao

Muitas vezes de posse de um conjuntos de dados ou pontos em um plano, nos questionamos se  possvel modelar uma curva que interprete o comportamento dessa distribuio de pontos e que passa por eles.

Na interpolao estudamos como modelar tais curvas de modo que essas condies so satisfeitas.




De maneira geral, temos que n pontos sero interpolados por um polinmio de grau n  1. A interpolao polinomial  um caso particular do problema geral de interpolao, no qual a famlia de funes  constituda de polinmios.

O teorema da Aproximao de Weierstrass, segundo o qual, qualquer funo contnua definida em um intervalo fechado pode ser aproximada uniformemente por um polinmio to bem quanto se queira.



O teorema de Weirestrass afirma que existe um polinmio to prximo quanto de queira da funo a ser interpolada. Para determinar esse polinmio precisamos de algumas tcnicas, uma delas  a resoluo de um sistema cujo sua montagem  determinada por yi = f(xi) para todo i  {1, 2, ..., n} e chamamos, essa tcnica, entre outros nomes de
<b>interpolao via matriz de Vandermonde</b>.

Tal matriz  dada por:



De forma que podemos afirmar que, para uma quantidade <b>n</b> de pontos, teremos um <b>polinmio de grau n-1</b> para representar nossa interpolao.
Para representar isso, vamos ver um exemplo:

<b>Sejam os pontos P1 = (1, 1), P2 = (2, 0.5), P3 = (3, 2) pertencentes  uma funo f(x). Encontre a funo de interpolao g(x) tais que g(x0) = f(x0) (condio mnima para curva ser interpolada).</b>

Para resolver esse exemplo, utilizaremos a biblioteca Numpy



```python
import numpy as np

x = np.array([1, 2, 3])
y = np.array([1, 0.5, 2])

A_matrix = np.vander(x) # Matriz para resolver a interpolao

b_matrix = y # Matriz das solues

solution = np.linalg.solve(A_matrix, b_matrix)

print(f'Matriz de Vandermonde\n {A_matrix}\n')
print(f'Matriz das f(x): {b_matrix}\n') 
print(f'Soluo dos coeficientes da curva interpolada: {solution}')

```
